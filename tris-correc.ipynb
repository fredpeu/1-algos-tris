# Introduction

On désigne par "tri" l'opération consistant à ordonner un ensemble d'éléments en fonction de clés sur lesquelles est définie une relation d'ordre.

Les algorithmes de tri ont une grande importance pratique. Ils sont fondamentaux dans certains domaines, comme l'informatique de gestion où l'on tri de manière quasi-systématique des données avant de les utiliser.

L'étude du tri est également intéressante en elle-même car il s'agit sans doute du domaine de l'algorithmique qui a été le plus étudié et qui a conduit à des résultats remarquables sur la construction d'algorithmes et l'étude de leur complexité.

Nous allons étudier ici deux types de tris:

* Le tri par sélection
* Le tri par insertion

# LE TRI PAR  `SELECTION`

> Observez  attentivement dans l´animation [algorithmes illustrés](http://www.fredpeuriere.com/NSI/algos-1/algorithmes-illustres/algos.html) le principe de fonctionnement du `tri par sélection`.

  ##  -Code Python:
  Ecrivez la fonction `triSelection(liste)` qui implémente cet algorithme en PYTHON. 

```python
def triSelection(liste):
    n=len(liste)
    for i in range(n-1):
        indice_min=i
        for j in range(i+1,n):
            if liste[j]<liste[indice_min]:
                indice_min=j
        liste[i],liste[indice_min]=liste[indice_min],liste[i]
    return liste
```

> Tester maintenant la fonction de tri par **sélection**:

```python
liste1=[9,8,7,6,5,4,3,2,1]
triSelection(liste1)
```

  ##  -Test avec des listes aléatoires:

> Ecrivez la fonction `rdmTab` qui prend la taille de la liste à créer en paramètre et retourne un liste de valeurs aléatoires comprises entre 0 et 1000. Testez ensuite la fonction en générant de **listes aléatoires** de différentes tailles:

```python
import random as rdm
def rdmTab(n):
    rdmList=[]
    for i in range(n):
        rdmList.append(rdm.randint(0,1000))
    return(rdmList)
liste2=rdmTab(20)
print(liste2)
```

> Testez le tri par sélection d´une **liste aléatoire** de 20 éléments:

```python
liste3=rdmTab(20)
print(triSelection(liste3))
```

  ##  -Tests de temps d´exécution:

La fonction de `tri par sélection` emploie deux boucles imbriquées. <br>On peut montrer facilement que le temps d´exécution <b> est proportionnel au carré du nombre de données à traiter</b> (n<sup>2</sup>).

> Montrez par l´exemple que le temps d´exécution du **tri par sélection** n´est pas proportionnel à la taille des données:

```python
from time import time

liste4=(rdmTab(1000))
liste5=(rdmTab(2000))

debut=time()
triSelection(liste4)
fin=time()
t1=fin-debut

debut=time()
triSelection(liste5)
fin=time()
t2=fin-debut

print(t1,t2,t2/t1)
# On voit que le temps d´exécution est environ 4 fois plus important si on double la taille des données.
```

# LE TRI PAR  `INSERTION`

> Observez  attentivement dans l´animation [algorithmes illustrés](http://www.fredpeuriere.com/NSI/algos-1/algorithmes-illustres/algos.html) le principe de fonctionnement du `tri par insertion`.

  ##  -Code Python:
  Ecrivez la fonction `triInsertion(liste)` qui implémente cet algorithme en PYTHON. 

```python
def triInsertion(liste):
    n=len(liste)
    for i in range (1,n):
        while ((i>0) and (liste[i]<liste[i-1])):
            liste[i-1],liste[i] = liste[i],liste[i-1]
            i=i-1
    return liste
```
